### Problem Statement : Visualize data with matplotlib and seaborn libraries by plotting graphs

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
raw_data = pd.read_csv("Heart.csv")
```

```python
raw_data.head()
```

```python
raw_data.describe()
```

```python
df = raw_data.iloc[:,1:]
df
```

```python
plt.figure(figsize=(12, 8))
plt.scatter("Age","MaxHR",data=df ,color= "red", marker =".")
plt.xlabel("Age")
plt.ylabel("MAxHR")
```

```python
plt.figure(figsize=(12, 8))

sns.lineplot(data=df, x="Age", y="Chol")
```

```python
plt.figure(figsize=(12, 8))

sns.histplot(data=df, x="Age", color="violet")
```

```python
plt.figure(figsize=(12, 8))
sns.boxplot(x="Age", y="RestBP", data=df)
```

```python
plt.figure(figsize=(12, 8))

sns.pairplot(df, diag_kind="hist")
```

```python
plt.figure(figsize=(12,8)) 
ax = sns.heatmap(df.corr(),cmap="YlGnBu", annot=True)
```

```python
plt.figure(figsize=(12, 8))
sns.lineplot(data=df, x="Age",y="ChestPain" ,hue="Sex")
```
