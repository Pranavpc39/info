### Data Transformation

```python
print(dataset.ChestPain.unique())
print(dataset.Thal.unique())
print(dataset.AHD.unique())
```

```python
heart_encoding = pd.get_dummies(dataset[['ChestPain', 'Thal', 'AHD']])
heart_final = pd.concat([dataset, heart_encoding],1)
heart_final = heart_final.drop(['ChestPain', 'Thal', 'AHD'], axis = 1)
heart_final.head(10)
```

```python
heart_final.Sex.value_counts()
```

```python
pd.crosstab(heart_final.AHD_Yes,heart_final.Sex)
```

```python
heart_final.columns
```

```python
heart_final.dtypes
```

```python
df = heart_final.drop('AHD_Yes', axis=1)
df_norm = (df-df.min())/(df.max()-df.min())
df_norm = pd.concat((df_norm, heart_final.AHD_Yes), 1)
```

```python
df_norm.head(10)
```

```python
X = df_norm.drop(['AHD_Yes', 'Unnamed: 0'], axis=1)
Y = df_norm.AHD_Yes
```

### Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.2 ,random_state=1)
```

### KNN Modelling

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
classifier = KNeighborsClassifier(n_neighbors=5, metric="minkowski", p=2)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix
conf_mat = confusion_matrix(y_test, y_pred)
sns.heatmap(conf_mat, square=True, annot=True, cmap='Blues', fmt='d', cbar=False)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

### Conclusion : Accuracy of trained model is 98.34%
